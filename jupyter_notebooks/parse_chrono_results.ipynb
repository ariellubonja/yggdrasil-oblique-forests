```python
def pivot_tree_timings(log_text: str) -> pd.DataFrame:
    """
    Return a (n_trees × n_functions) table:
      • index  → tree number
      • columns → function names
      • values  → summed seconds spent in that function for that tree
    """
    df = parse_yggdrasil_timings(log_text)

    # Keep only lines where we actually know which tree they belong to
    df = df[df["tree"].notna()]

    pivot = (
        df.groupby(["tree", "function"], as_index=False)["time_s"]
          .sum()                       # sum if same function appears several times
          .pivot(index="tree", columns="function", values="time_s")
          .fillna(0.0)                 # trees may lack some functions
    )

    # nicer formatting (optional)
    pivot.columns.name = None
    pivot.index.name = "tree"
    return pivot
```

```python
import re
import pandas as pd


# ── 1. Regex for “xxx took: 0.123s”  (case-insensitive) ──────────────
TIMING_RE = re.compile(
    r"^\s*(?:-\s*)*"          # any leading “ - ” blocks
    r"(?P<name>[^:]+?)\s+"    # function name (lazy up to colon)
    r"(?:Took|took):\s+"      # “Took:” OR “took:”  ← key change
    r"(?P<secs>[0-9.eE+-]+)s",
    re.IGNORECASE,
)


# ── 2. Scan the log and tag each timing with its tree number ──────────
def _parse(log: str) -> pd.DataFrame:
    rows, cur_tree = [], -1
    for line in log.splitlines():
        if "Selecting Bootstrapped Samples Took" in line:
            cur_tree += 1                      # next tree starts
        m = TIMING_RE.match(line)
        if m and cur_tree >= 0:                # ignore pre-tree garbage
            rows.append(
                {"tree": cur_tree + 1,         # 1-based index
                 "function": m.group("name").strip(),
                 "time_s": float(m.group("secs"))}
            )
    return pd.DataFrame(rows)


# ── 3. Pivot into a wide “one-row-per-tree” table ─────────────────────
def tree_table(log: str) -> pd.DataFrame:
    df = _parse(log)
    long = df.groupby(["tree", "function"], as_index=False)["time_s"].sum()
    wide = (long
            .pivot(index="tree", columns="function", values="time_s")
            .fillna(0.0)
            .sort_index(axis=1))
    wide.index.name = "tree"
    wide.columns.name = None
    return wide
```
